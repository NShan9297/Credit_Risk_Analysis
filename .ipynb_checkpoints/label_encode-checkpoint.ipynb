```python
import pandas as pd
from pathlib import Path

file_path = Path("./Resources/loans_data.csv")
loans_df = pd.read_csv(file_path)
loans_df.head()
```

```python
#the target is the loan status which is either bad or not bad
```

```python
#encoding the gender columns to numbers--0 is false, 1 is true
loans_binary_encoded = pd.get_dummies(loans_df, columns=["gender"])
loans_binary_encoded.head()
```

```python
#possible to encode multiple columns at a time
loans_binary_encoded = pd.get_dummies(loans_df, columns=["education", "gender"])
loans_binary_encoded.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df2 = loans_df.copy()
df2['education'] = le.fit_transform(df2['education'])
print(df2)
```

## Creating Custom Encoding

```python
label_encoder=LabelEncoder()
loans_df['month_le']=label_encoder.fit_transform(loans_df['month'])
loans_df.head()
```

```python
#Creating a dictionary of the months of the year and apply a custom function to convert the month names
#to their corresponding integers
```

```python
months_num = {
   "January": 1,
   "February": 2,
   "March": 3,
   "April": 4,
   "May": 5,
   "June": 6,
   "July": 7,
   "August": 8,
   "September": 9,
   "October": 10,
   "November": 11,
   "December": 12,
}
```

```python
#lambda funciton applied to the month column to perform the conversion
loans_df['month_num']=loans_df['month'].apply(lambda x: months_num[x])
```

```python
loans_df.head()
```

```python
#clean up- dropping the unnecessary columns related to the month 
loans_df = loans_df.drop(["month", "month_le"], axis=1)
loans_df.head()
```
